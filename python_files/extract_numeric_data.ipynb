```python
# import things
import pandas as pd
```

```python
# read txt file
path = "../text_files/dec.txt"
with open(path, 'r') as f:
    data = f.read()
    # save data to a list by line by line
    data = data.splitlines()
```

```python
print(data[0])
# split data by space
data = [i.split() for i in data]
print(data[0])
```

## Cordinate Extraction

```python
def get_label(data): # retrieve only 1 item from the list
    labels_list = ['Unknown' , 'Stellar' , 'Cluster' , 'Galaxy' , 'Chip']
    for i in range(len(data)):
        if data[i] in labels_list:
            if data[i] == 'Chip':
                return 'Chip Gap' # convert to original label
            return data[i]


def get_numeric_data(data): # retrieve only 1 item from the list
    data_list = data[1:7]
    data_list = [float(i) for i in data_list]
    return data_list
```

```python
# extract numeric data and labels from data
for i in data:
    # print(i)
    print(get_label(i), get_numeric_data(i))
    # break
```

```python
# creat pandas dataframe containing numeric data from get_numeric_data and labels from get_label
df = pd.DataFrame([get_numeric_data(i) for i in data], columns=['A', 'B', 'C', 'D', 'E', 'F'])
df['Class'] = [get_label(i) for i in data]
print(df.head())
```

## Shape & Light Extraction

```python
# read txt file
path = "../text_files/M33 Adopted Cluster Catalog.txt"
with open(path, 'r') as f:
    data_num = f.read()
    # save data to a list by line by line
    data_num = data_num.splitlines()
```

```python
# strip space from first and last
data_num = [i.strip() for i in data_num]
```

```python
data_num
```

```python
# remove last items ultil the last item is in ['Unknown' , 'Stellar' , 'Cluster' , 'Galaxy' , 'Chip']
for i in range(len(data_num)):
    while data_num[i].split(' ')[-1] not in ['Unknown' , 'Stellar' , 'Cluster' , 'Galaxy' , 'Chip']:
        data_num[i] = data_num[i].rsplit(' ', 1)[0]
```

```python
# replace 'Â· Â· Â·' with 'missing value'
data_num = [i.replace('Â· Â· Â·', 'missing_value') for i in data_num]
```

```python
# split data by tab
data_num = [i.split(' ') for i in data_num]
```

```python
data_num = [data_num[i][3:] for i in range(len(data_num))]
```
