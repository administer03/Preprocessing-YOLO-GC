```python
path_img_train = '../../data/ds9_dataset/full_img_flip/train/*'
path_img_test = '../../data/ds9_dataset/full_img_flip/test/*'
```

```python
# get all folder name
folder_name_train = glob.glob(path_img_train)
folder_name_test = glob.glob(path_img_test)
```

```python
# display image x with original 
img = cv2.imread(x, cv2.IMREAD_GRAYSCALE)
# flip y axis
img_flip = cv2.flip(img, 0)
# display image side by side
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.title('Flip Y-axis')
plt.imshow(img_flip, cmap='gray')
plt.show()
```

# Export images

```python
def save_image(sub_filter_path):
    # get all image path
    img_path = glob.glob(sub_filter_path + '/*.jpg')
    # loop through all image path
    for x in img_path:
        # read image
        img = cv2.imread(x, cv2.IMREAD_GRAYSCALE)
        # flip y axis
        img_flip = cv2.flip(img, 0)
        # save image with 75% quality
        cv2.imwrite(x, img_flip, [cv2.IMWRITE_JPEG_QUALITY, 75])
```

```python
# loop through all folder name with tqdm
for x in tqdm.tqdm(folder_name_train):
    save_image(x)

for x in tqdm.tqdm(folder_name_test):
    save_image(x)
```
